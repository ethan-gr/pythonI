# Formato GenBank

```python
from Bio import SeqIO
file = '../data/virus.gb'
```

```python
# Ejercicio 4: Obtener información de las anotaciones
# Utilizando el archivo virus.gb obtengan de qué organismo viene y la fecha.
for gb_record in SeqIO.parse(file, "genbank"):
    print(f"Organismo: {gb_record.annotations['organism']}")
    print(f"fecha:{gb_record.annotations['date']}")
```

```python
# Ejercicio 5: Obtener información de las features
# Utilizando el archivo virus.gb obtengan el número del aislado y el país de donde proviene
for gb_record in SeqIO.parse(file, "genbank"):
    print(f"País: {gb_record.features[0].qualifiers['country'][0]}")
    print(f"Aislado: {gb_record.features[0].qualifiers['isolation_source'][0]}")
```

```python
# Ejercicio 6
# Utilizando el archivo virus.gb obtengan la secuencia, transcripción y traducción del gen Lfor gb_record in SeqIO.parse(file, "genbank"):
for gb_record in SeqIO.parse(file, "genbank"):
    for feature in gb_record.features:
        if 'gene' and 'translation' in feature.qualifiers.keys() and feature.qualifiers['gene'][0] == 'L':
            seq = gb_record.seq[feature.location.nofuzzy_start:feature.location.nofuzzy_end]
            print(f"Gen: {feature.qualifiers['gene'][0]}")
            print(f"ADN: {seq}")
            print(f"ARN: {seq.transcribe()}")
            print(f"Proteína: {feature.qualifiers['translation'][0]}") # | print(seq.translate())
```
